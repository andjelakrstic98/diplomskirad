```python
import matplotlib.pyplot as plt

tastature = [6,5,2]
oznake = 'Redragon','Trust','Nijedna'
boje = ['mediumspringgreen','violet','coral']
my_explode = (0.1,0,0)

plt.pie(tastature,labels=oznake,autopct='%1.1f%%', startangle=15, shadow = True, colors=boje, explode=my_explode)
plt.title('Odgovori na pitanje u procentima')
plt.axis('equal')
plt.show()
```
